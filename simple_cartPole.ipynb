```
# import all libs
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
!pip install gym
import gym
```

FURTHER CODE IS DIVIDED INTO THREE SECTIONS IMPORTING DATA(WITH FILTER) AND DEPENDENCIES ,THEN MODEL DEFINITION & ATLAST PREDICTIONS

```
#############################################
# SIMPLE KERAS MODEL FOR OPENAI GYM  ########
#############################################

# GATHER DATA FROM GIVEN ENV AND PUT A FILTER ON IT
def gather_data(env):
  min_score=50  #threshold filter value
  sim_steps=500 #simple steps
  trainingX,trainingY=[],[]  #2 arrs containig training sets
  scores=[]  #arr for storing scores
  for _ in range(10000):
#     take new observation
    obs=env.reset()
#   initialize score
    score=0
#   pre-training samples arrs X,Y
    training_sampleX,training_sampleY=[],[]
#  
    for step in range(sim_steps):
      action=np.random.randint(0,2) #random value between 0 to 2
      one_hot_action=np.zeros(2)  #initiate an arr of ([0.0,0.0])
      one_hot_action[action]=1    #value of random element in arr one_hot_action is 1
      training_sampleX.append(obs) #Append observations in X
      training_sampleY.append(one_hot_action) #append actions in Y

      observations,reward,done,_=env.step(action) #store values of obs,reward received and whether process is complted or not by taking any random action
      score=score+reward    #add reward in score and store the score
      if done:              # if process is completed then break else continue..........
         break

    if score>min_score:    #check for every iterationwhether score is higher than 50 if yes then only....
      scores.append(score)  #append that score to scores[] array
      trainingX=trainingX+training_sampleX
      trainingY=trainingY+training_sampleY
  
  
  trainingX,trainingY= np.array(trainingX),np.array(trainingY)  #convert values for training into arrays
  print("average: {}".format(np.mean(scores))) #mean of scores[]
  print("median: {}".format(np.median(scores))) #median of scores[]
  return trainingX,trainingY

# MODEL DEFINITION
# A CNN TYPE MODEL IS BEING UTILIZED FOR THIS CASE

def create_model():
    model = Sequential()
    model.add(Dense(128, input_shape=(4,), activation="relu"))
    model.add(Dropout(0.6))
    
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.6))
    
    model.add(Dense(512, activation="relu"))
    model.add(Dropout(0.6))
    
    model.add(Dense(256, activation="relu"))
    model.add(Dropout(0.6))
    
    model.add(Dense(128, activation="relu"))
    model.add(Dropout(0.6))
    model.add(Dense(2, activation="softmax"))
    
    model.compile(
        loss="categorical_crossentropy",
        optimizer="adam",
        metrics=["accuracy"])
    return model
  
#   PREDICTION 
def predict():
  env=gym.make("CartPole-v1")   #we r training on cartpole model of openai gym envs
  trainingX,trainingY=gather_data(env)   #gather required data from env
  model=create_model()               #import our CNN-model to model variable
  model.fit(trainingX,trainingY,epochs=5)  #fit our model into datasets for 5 epochs 
#   self explainatory
  scores=[]
  num_trials=50
  sim_steps=500
  
  for trial in range(num_trials):
#     self-explainatory
    score=0
    obs=env.reset()   #take new observations
    for step in range(sim_steps):
      action = np.argmax(model.predict(obs.reshape(1,4))) #Returns indices of the max element of the array in a particular axis.
      observation, reward, done, _ = env.step(action)
      score += reward
      if done:
        break
    scores.append(score)
    
  print(np.mean(scores))
  
  
if __name__=="__main__":
  predict()
  
```
